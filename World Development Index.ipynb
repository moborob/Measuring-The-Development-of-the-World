```python
data2 = data.set_index(["Country Name"])
data2
```

```python
#what will be utilized for project:

# 1. Population, total 
# 2. Access to electricity (% of population)
# 3. Children out of school, primary 
# 4. CO2 emissions (kt) 
# 5. Death rate, crude (per 1,000 people) 
# 6. GDP per capita (current US$) 
# 7. Life expectancy at birth, total (years) 
# 8. Literacy rate, adult total (% of people ages 15 and above) 
# 9. People using at least basic drinking water services (% of population) 
# 10. People using at least basic sanitation services (% of population) 
# 11. Poverty gap at $5.50 a day (2011 PPP) (%)  
# 12. School enrollment, primary (% net) 
# 13. School enrollment, secondary (% net) 
```

```python
# by region: 
#what will be used:
CSS = data2[data2['Country Code']=='CSS'] #* carribean small states 
EAS = data2[data2['Country Code']=='EAS'] #* East Asia and Pacific
ECS= data2[data2['Country Code']=='ECS'] #* Europe and central asia
LCN= data2[data2['Country Code']=='LCN'] #* Latin America and Carribean
MEA = data2[data2['Country Code']=='MEA'] #* Middle East and North Africa
NAC = data2[data2['Country Code']=='NAC'] #* North America
SAS = data2[data2['Country Code']=='SAS']#* South Asia
SSF = data2[data2['Country Code']=='SSF'] #* Sub-saharan Africa
```

```python
concatenated_data = pd.concat([CSS, EAS, ECS, LCN, MEA, NAC, SAS, SSF])
concatenated_data
```

```python
population = concatenated_data[concatenated_data["Indicator Name"] == "Population, total"]
population
```

```python
population_T = population.T
population_data = population_T[3:]
population_data
```

```python
population_data.dropna().plot()
plt.xlabel("years")
plt.ylabel("population (in billions)")
plt.title("Population by Region")
plt.legend(loc = "best", labelspacing=0.1)
#legend(..., prop={'size':8}) for reference
```

```python
population_total = population_data.dropna().sum(axis = 1)
population_total.plot()
plt.xlabel("years")
plt.ylabel("population (in billions)")
plt.title("Total Population")
```

```python
#Access to electricity (% of population)
electric = concatenated_data[concatenated_data["Indicator Name"] == "Access to electricity (% of population)"]
```

```python
#Transposing and filtering data
electric_T = electric.T
electric_data = electric_T[3:]
```

```python
#data visualization for electricity accessibility
electric_data.dropna().plot()
plt.xlabel("years")
plt.ylabel("% of people")
plt.title("Percent of people with access to electricity by region")
plt.legend(loc = "best", labelspacing=0.1, prop={'size':7})
#legend(..., prop={'size':8}) for reference
```

```python
#avergae percent of access to electricity by region
electric_data_total= electric_data.dropna().mean(axis = 1)
electric_data_total.plot()
plt.xlabel("years")
plt.ylabel("% of people")
plt.title("Percent of people with access to electricity in the world")
```
